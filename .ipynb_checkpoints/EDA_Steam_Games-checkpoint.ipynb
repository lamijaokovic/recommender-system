# Normalization & Scoring logic

### Generate new column called ```normalized_rating``` which is taking ```positive_ratings``` and ```negative_ratings``` and it calculates the normalized rating using a weighted average with a smoothing factor ```α (alpha)``` .

````python
alpha = 10
df['normalized_rating'] = ( (df['positive_ratings'] + alpha) / (df['positive_ratings'] + df['negative_ratings'] + 2 * alpha) )
````

### Test to show games with 0 playtime

````python
#normalizedaverage_playtime = np.log1p(df['average_playtime'] - df['average_playtime'].min()) / (df['average_playtime'].max() - df['average_playtime'].min())

#normalizacija = np.log1p(df['average_playtime'])/10
#normalizacija

#df['average_playtime']

#games_with_zero_playtime_names = df.loc[df['average_playtime'] == 0, 'name'].tolist()
#print("Igre s nulom u average_playtime:", games_with_zero_playtime_names)

count_games_with_zero_playtime = df['average_playtime'].eq(0).sum()
print("Broj igara s nulom u average_playtime:", count_games_with_zero_playtime)

games_with_zero_playtime = df[df['average_playtime'] == 0]
print(games_with_zero_playtime[['name']])
````

### Generate new column called ```owners_score``` which will return score based on rules we set. Minimum score is ```0.1``` and maximum score is ```1```




````python
def getScoreByOwnersCount(ownersCount):
  if ownersCount == "0-20000":
    return 0.1
  elif ownersCount == "20000-50000":
    return 0.2
  elif ownersCount == "50000-100000":
    return 0.3
  elif ownersCount == "100000-200000":
    return 0.4
  elif ownersCount == "200000-500000":
    return 0.5
  elif ownersCount == "500000-1000000":
    return 0.6
  elif ownersCount == "1000000-2000000":
    return 0.7
  elif ownersCount == "2000000-5000000":
    return 0.75
  elif ownersCount == "5000000-10000000":
    return 0.80
  elif ownersCount == "10000000-20000000":
    return 0.85
  elif ownersCount == "20000000-50000000":
    return 0.90
  elif ownersCount == "50000000-100000000":
    return 0.95
  else:
    return 1

df['owners_score'] = df['owners'].apply(lambda x: getScoreByOwnersCount(x))
````

### Generate new column called ```normalized_playtime``` which is taking games  ```average_playtime``` and sum of all ```average_playtime```

````python
totalPlayTime = df['average_playtime'].sum()

def getNormalizedPlaytime(playtime):
  return np.log1p(playtime) / np.log1p(totalPlayTime)


df['normalized_playtime'] = df['average_playtime'].apply(lambda x: getNormalizedPlaytime(x))
````

### Generate new columns called ```time_engaged_index```, ```player_acclaim_index```, ```adopters_choice_index``` and ```game_nexus_index```

*   ```time_engaged_index``` will have its main focus on ```normalized_playtime``` parameter, but it will also consider the ```normalized_rating``` and ```owners_score``` but they will have significantly less impact to final score.

*   ```player_acclaim_index``` will have its main focus on ```normalized_rating``` parameter, but it will also consider the ```normalized_playtime``` and ```owners_score``` but they will have significantly less impact to final score.

*   ```adopters_choice_index``` will have its main focus on ```owners_score``` parameter, but it will also consider the ```normalized_rating``` and ```normalized_playtime``` but they will have significantly less impact to final score.

*   ```game_nexus_index``` will take all parameters with same importance.

````python
df['time_engaged_index'] = (0.05 * df['normalized_rating']) + (0.9 * df['normalized_playtime']) + (0.05 * df['owners_score'])
df['player_acclaim_index'] = (0.9 * df['normalized_rating']) + (0.05 * df['normalized_playtime']) + (0.05 * df['owners_score'])
df['adopters_choice_index'] = (0.05 * df['normalized_rating']) + (0.05 * df['normalized_playtime']) + (0.9 * df['owners_score'])
df['game_nexus_index'] = (0.33 * df['normalized_rating']) + (0.33 * df['normalized_playtime']) + (0.33 * df['owners_score'])
````

### Optimize ```game_nexus_index``` by using ```np.percentile``` to determine 10 score categories. These categories will be later used in vectorization process for KNN recommender.

````python
# Calculate percentiles for each category
nexusPercentiles = np.percentile(df['game_nexus_index'], np.linspace(0, 100, 11))
adoptersChoicePercentiles = np.percentile(df['adopters_choice_index'], np.linspace(0, 100, 11))
playerAcclaimPercentiles = np.percentile(df['player_acclaim_index'], np.linspace(0, 100, 11))
timeEngagedPercentiles = np.percentile(df['time_engaged_index'], np.linspace(0, 100, 11))

# Display the numerical range for each category
for i in range(10):
  print(f"Nexus Category {i + 1}: {nexusPercentiles[i]:.3f} to {nexusPercentiles[i + 1]:.3f}")
print("")

for i in range(10):
  print(f"Adopter Category {i + 1}: {adoptersChoicePercentiles[i]:.3f} to {adoptersChoicePercentiles[i + 1]:.3f}")
print("")

for i in range(10):
  print(f"Player Accl. Category {i + 1}: {playerAcclaimPercentiles[i]:.3f} to {playerAcclaimPercentiles[i + 1]:.3f}")
print("")

for i in range(10):
  print(f"Time Eng. Category {i + 1}: {timeEngagedPercentiles[i]:.3f} to {timeEngagedPercentiles[i + 1]:.3f}")
````

### Use generated categories from above to generate string values depending on score

````python
#Get Nexus Category
def getNexusCategory(gameNexusIndex):
  if gameNexusIndex <= 0.191:
    return "Unacceptable"
  elif gameNexusIndex <= 0.206:
    return "Subpar"
  elif gameNexusIndex <= 0.220:
    return "Mediocre"
  elif gameNexusIndex <= 0.235:
    return "BelowAverage"
  elif gameNexusIndex <= 0.253:
    return "Average"
  elif gameNexusIndex <= 0.277:
    return "AboveAverage"
  elif gameNexusIndex <= 0.309:
    return "Good"
  elif gameNexusIndex <= 0.375:
    return "Excellent"
  elif gameNexusIndex <= 0.488:
    return "Outstanding"
  else:
    return "Masterpiece"

df['game_nexus_category'] = df['game_nexus_index'].apply(lambda x: getNexusCategory(x))

#Get Adopter Choice Category
def getAdoptersChoiceCategory(adoptersChoiceIndex):
  if adoptersChoiceIndex <= 0.114:
    return "Unacceptable"
  elif adoptersChoiceIndex <= 0.116:
    return "Subpar"
  elif adoptersChoiceIndex <= 0.118:
    return "Mediocre"
  elif adoptersChoiceIndex <= 0.121:
    return "BelowAverage"
  elif adoptersChoiceIndex <= 0.124:
    return "Average"
  elif adoptersChoiceIndex <= 0.129:
    return "AboveAverage"
  elif adoptersChoiceIndex <= 0.206:
    return "Good"
  elif adoptersChoiceIndex <= 0.282:
    return "Excellent"
  elif adoptersChoiceIndex <= 0.418:
    return "Outstanding"
  else:
    return "Masterpiece"

df['adopters_choice_category'] = df['adopters_choice_index'].apply(lambda x: getAdoptersChoiceCategory(x))

#Get Player Acclaim Category
def getPlayerAcclaimCategory(playerAcclaimIndex):
  if playerAcclaimIndex <= 0.427:
    return "Unacceptable"
  elif playerAcclaimIndex <= 0.472:
    return "Subpar"
  elif playerAcclaimIndex <= 0.496:
    return "Mediocre"
  elif playerAcclaimIndex <= 0.530:
    return "BelowAverage"
  elif playerAcclaimIndex <= 0.570:
    return "Average"
  elif playerAcclaimIndex <= 0.613:
    return "AboveAverage"
  elif playerAcclaimIndex <= 0.662:
    return "Good"
  elif playerAcclaimIndex <= 0.718:
    return "Excellent"
  elif playerAcclaimIndex <= 0.791:
    return "Outstanding"
  else:
    return "Masterpiece"

df['player_acclaim_category'] = df['player_acclaim_index'].apply(lambda x: getPlayerAcclaimCategory(x))

#Get Time Engage Category
def getTimeEngageCategory(timeEngageIndex):
  if timeEngageIndex <= 0.029:
    return "Unacceptable"
  elif timeEngageIndex <= 0.031:
    return "Subpar"
  elif timeEngageIndex <= 0.033:
    return "Mediocre"
  elif timeEngageIndex <= 0.036:
    return "BelowAverage"
  elif timeEngageIndex <= 0.039:
    return "Average"
  elif timeEngageIndex <= 0.042:
    return "AboveAverage"
  elif timeEngageIndex <= 0.048:
    return "Good"
  elif timeEngageIndex <= 0.222:
    return "Excellent"
  elif timeEngageIndex <= 0.377:
    return "Outstanding"
  else:
    return "Masterpiece"

df['time_engaged_category'] = df['time_engaged_index'].apply(lambda x: getTimeEngageCategory(x))
````
